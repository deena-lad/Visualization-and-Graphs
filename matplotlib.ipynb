## Matplotlib library

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Data for plotting

```python
# 100 linearly spaced numbers from 0 to 10
x = np.linspace(0, 10, 100)  
# Sine of x
y = np.sin(x)        
# Cosine of x
y2 = np.cos(x)               
```

### Line plot

```python
plt.figure(figsize=(10, 5)) 
plt.plot(y, label='Sine', color='blue')  
plt.plot(y2, label='Cosine', color='red')           
plt.title('Line Plot of Sine and Cosine Functions')                    
plt.xlabel('X Axis')                                                   
plt.ylabel('Y Axis')                                                  
plt.legend()                                                          
plt.grid(True)                                                        
plt.show()
```

### Scatter plot

```python
plt.figure(figsize=(10, 5))  
plt.scatter(x, y, label='Sine', color='blue', marker='o')             
plt.scatter(x, y2, label='Cosine', color='red', marker='x')            
plt.title('Scatter Plot of Sine and Cosine Functions')                
plt.xlabel('X Axis')                                                  
plt.ylabel('Y Axis')                                                   
plt.legend()                                                           
plt.grid(True)                                                        
plt.show()                                                             
```

### Bar plot

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]
```

```python
plt.figure(figsize=(7, 4))  
plt.bar(categories, values, color='orange')                            
plt.title('Bar Plot Example')                                        
plt.xlabel('Categories')                                             
plt.ylabel('Values')                                                
plt.show()
```

### Histogram

```python
data = np.random.randn(1000)  
```

```python
plt.figure(figsize=(10, 5))  
plt.hist(data, bins=30, color='purple', edgecolor='black')           
plt.title('Histogram Example')                                     
plt.xlabel('Value')                                                   
plt.ylabel('Frequency')                                             
plt.show()
```

### Pie chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
```

```python
plt.figure(figsize=(5, 5)) 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)  
plt.title('Pie Chart Example')                                     
plt.axis('equal')                                                     
plt.show()
```

### Box plot

```python
data = [np.random.randn(100) for _ in range(4)]  
```

```python
plt.figure(figsize=(10, 5)) 
plt.boxplot(data, patch_artist=True, vert=True, labels=['A', 'B', 'C', 'D'])  
plt.title('Box Plot Example')                                               
plt.xlabel('Category')                                                     
plt.ylabel('Value')                                                          
plt.show()  
```

### Heatmap

```python
matrix = np.random.rand(10, 10)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(matrix, cmap='viridis', interpolation='none')                      
plt.title('Heatmap Example')                                                 
plt.colorbar()                                                               
plt.show()                                                                     
```
