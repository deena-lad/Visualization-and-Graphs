```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Setting Seaborn style for all plots
sns.set()
```

### Data

```python
# Sample data
np.random.seed(42)
data = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]
```

```python
# Creating a DataFrame for categorical plots
df = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Value': np.random.randn(100)
})
```

### Scatter plot

```python
tips = sns.load_dataset("tips")  # Load example dataset
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", style="time", palette="deep")
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Line plot

```python
plt.figure(figsize=(12, 7))
sns.lineplot(x="total_bill", y="tip", data=tips, hue="day", style="time", markers=True, dashes=False)
plt.title('Line Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Bar plot

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=categories, y=values, palette="muted")
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### Histogram 

```python
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color='blue')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Box plot

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="Category", y="Value", data=df, palette="coolwarm")
plt.title('Box Plot Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Violin Plot

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x="Category", y="Value", data=df, palette="Set2")
plt.title('Violin Plot Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Heatmap

```python
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap Example')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

### Pair plot

```python
iris = sns.load_dataset("iris")
```

```python
plt.figure()
sns.pairplot(iris, hue="species", palette="bright")
plt.suptitle('Pair Plot Example', y=1.02)
plt.show()
```

### Count Plot

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="Category", data=df, palette="Set1")
plt.title('Count Plot Example')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
